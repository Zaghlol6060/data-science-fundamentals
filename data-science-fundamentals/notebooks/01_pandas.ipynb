# 📊 Pandas Series

## 🧠 ما هي Series؟

الـ **Series في Pandas** هي عبارة عن **مصفوفة أحادية البعد (1D)** تحتوي على بيانات، وكل عنصر فيها له **Label (Index)**.

يمكنك تخيلها كأنها **عمود واحد في جدول Excel**.

---

## 🔑 الخصائص الأساسية

- أحادية البعد (1D)
- تحتوي على **Index (تسميات)** لكل عنصر
- يمكن أن تخزن أي نوع من البيانات (أرقام، نصوص، ... إلخ)
- تشبه قائمة (List) ولكن مع Labels


````python
data = [100,102,104,200,300]
````

````python
series = pd.Series(data)
print(series) # Create a Pandas Series from a list and display it
````

````python
series = pd.Series(data,index=["a","b","c","d","e"])
print(series)                 
````


في المثال ده، قمنا بإنشاء Series وحددنا **Index خاص** بدل الأرقام الافتراضية

بدل ما يكون الـ index:
0, 1, 2, 3, 4  

بقوا:
a, b, c, d, e  

👉 كل قيمة بقت مرتبطة باسم (label) بدل رقم، وده بيسهل الوصول للبيانات والتعامل معاها.

````python
print(series.loc["c"])  # الوصول لقيمة العنصر اللي الـ index بتاعه "c"

print("-"*20)

series.loc["a"] = 110   # تعديل قيمة العنصر اللي الـ index بتاعه "a"

print(series)  # عرض الـ Series بعد التعديل
````


👉 هنا بيعرض الـ Series بعد التعديل:

- القيمة عند index = "a" اتغيرت من 100 إلى 110  
- باقي القيم زي ما هي  
- `dtype: int64` يعني نوع البيانات أرقام صحيحة  

---

💡 الخلاصة:
- قدرنا نوصل لقيمة باستخدام `loc`  
- وعدلنا قيمة عنصر باستخدام الـ index

##### 📅 Tracking Daily Calories

لو عايز أتابع نظامي الغذائي، ممكن أسجل **عدد السعرات لكل يوم** باستخدام Dictionary.

##### 💡 الفكرة

- Key → اليوم  
- Value → إجمالي السعرات في اليوم  

````python
calories = {"Day 1":1750,"Day 2":2100,"Day 3":1700,"Day 4":2500,"Day 5":30000}
series_calorie = pd.Series(calories)
print(series_calorie)
````

````python
# عايز اشوف يومي الاول اكلت بمقدار قد ايه 
print("Day1", series_calorie.loc["Day 1"])

print("-"*20)

#يوم اكلت فيه كوكيز او حاجه زياده 
series_calorie.loc["Day 2"]+=500
print(series_calorie.loc["Day 2"])

print("-"*20)

#عايز اعرف الايام الي اكلت فيها زياده عن المطلوب 
print(series_calorie[series_calorie >2000])
````

## Pandas - Selection by Row

في مكتبة Pandas، يمكننا اختيار الصفوف (Rows) من الـ DataFrame باستخدام أكثر من طريقة مثل `loc` و `iloc`.

هذه العملية تُستخدم عندما نحتاج للوصول إلى بيانات صف معين أو مجموعة من الصفوف بناءً على الفهرس (Index) أو الموقع الرقمي.

وتُعتبر من الأساسيات المهمة في التعامل مع البيانات وتحليلها.

````python
print(df.loc[0]) # عايز اول row 
````

````python
df= pd.read_csv("data_banda.csv", index_col = "Name")
print(df)
````

When using `index_col="Name"` inside `read_csv`, we specify that the "Name" column will be used as the index of the DataFrame instead of the default numeric index (0, 1, 2, ...).

This helps in accessing the data in a more organized and easier way by using labels (names) instead of numbers.

````python
print(df.loc["Bulbasaur"])  # الآن يتم البحث باستخدام الاسم (Index) بدل الأرقام
````

````python
print(df.loc["Bulbasaur",["Height","Weight"]]) # اختيار صف "Bulbasaur" مع تحديد أعمدة معينة (Height و Weight) فقط
````

````python
print(df.loc["Bulbasaur":"Dratini",["Height","Weight"]]) # اختيار مجموعة صفوف من "Bulbasaur" إلى "Dratini" (شاملين) مع عرض أعمدة Height و Weight فقط
````

````python
print(df.iloc[0:11])  # من الصف الأول إلى الصف 11 
````

````python
print(df.iloc[0:11:2])
# iloc[start:end:step]
# هنا بنختار من الصف الأول إلى الصف 11 مع تخطي صف (step = 2)
````

````python
print(df.iloc[0:11:2 , 0:3])
# iloc[row_start:row_end:step , col_start:col_end]
# صفوف: من أول صف إلى الصف 11 مع تخطي صف (step = 2)
# أعمدة: من العمود 0 إلى 2 (العمود 3 غير مشمول)
````

````python
pokemon = input("enter a pokemon name: ")  
# طلب إدخال اسم من المستخدم

try:
    print(df.loc[pokemon])  
    # محاولة عرض عمود باسم المدخل
except KeyError:
    print(f"{pokemon} not found")  
    # في حالة عدم وجود الاسم، يتم طباعة رسالة خطأ
````

##### Try & Except

- `try`: نضع بداخلها الكود المتوقع حدوث خطأ فيه.
- `except`: يتم تنفيذها إذا حدث الخطأ.

في هذا المثال:
إذا كان الاسم غير موجود في الـ DataFrame، سيحدث خطأ من نوع `KeyError`،  
وبدلاً من توقف البرنامج، يتم طباعة رسالة توضح أن الاسم غير موجود.

## Filtering in Pandas

Filtering means selecting a subset of rows from a DataFrame based on a specific condition.

In simple words, it is **keeping the rows that match a condition** and removing the rest.

This helps us focus only on the data we need for analysis.

````python
Tall_pokemon = df[df["Height"] >=2]
print(Tall_pokemon)
````

````python
Heavy_pokemon = df[df["Weight"] >100]
print(Heavy_pokemon)
````

````python
legenDary_pokemon = df[df["Legendary"] == 1]
print(legenDary_pokemon)
````

````python
Water_pokemon = df[(df["Type1"] == "Water") | 
                   (df["Type2"] == "Water") ]

print(Water_pokemon)  
# اختيار البوكيمونات التي يكون نوعها Water سواء في Type1 أو Type2
````

````python
ff_pokemon = df[(df["Type1"] == "Fire") &
                (df["Type2"] == "Flying") ]

print(ff_pokemon)  
# اختيار البوكيمونات التي نوعها Fire في Type1 و Flying في Type2
````

## Aggregation in Pandas

Aggregation means reducing a set of values into a single summary value.

It is used to summarize and analyze data, such as calculating:
- sum
- mean
- count
- max
- min

Aggregation is often used with the `groupby()` function to analyze data by groups.

````python
df=pd.read_csv("data_banda.csv") 
print(df)
````

***************************

###### whole dataframe 
استخدام `numeric_only=True` مهم لأنه يمنع حدوث أخطاء عند وجود أعمدة نصية (مثل الأسماء أو الأنواع).

````python
print("=== MEAN ===")
print(df.mean(numeric_only=True), "\n")

print("=== SUM ===")
print(df.sum(numeric_only=True), "\n")

print("=== MIN ===")
print(df.min(numeric_only=True), "\n")

print("=== MAX ===")
print(df.max(numeric_only=True), "\n")

print("=== COUNT ===")
print(df.count(numeric_only=True))
````

**********************************

##### Aggregation - Single Column

في هذا الجزء نقوم بتطبيق عمليات الـ Aggregation على عمود واحد فقط بدل الـ DataFrame بالكامل.

````python
print("=== HEIGHT MEAN ===")
print(df["Height"].mean(), "\n")

print("=== HEIGHT SUM ===")
print(df["Height"].sum(), "\n")

print("=== HEIGHT MIN ===")
print(df["Height"].min(), "\n")

print("=== HEIGHT MAX ===")
print(df["Height"].max(), "\n")

print("=== TYPE2 COUNT ===")
print(df["Type2"].count())
````

*******************************

````python
group = df.groupby("Type1")
print(group)
````

لأن `groupby()` لا تقوم بإظهار البيانات مباشرة.

هي فقط تقوم بإنشاء كائن (Object) يحتوي على مجموعات البيانات،  
لكن لم يتم بعد تطبيق أي عملية مثل العرض أو الحساب.

````python
print("=== MEAN ===")
print(group["Height"].mean(), "\n")

print("=== SUM ===")
print(group["Height"].sum(), "\n")

print("=== MIN ===")
print(group["Height"].min(), "\n")

print("=== MAX ===")
print(group["Height"].max(), "\n")

print("=== COUNT ===")
print(group["Height"].count())
````

لو عندك بوكيمونات:
- Fire
- Water
- Grass

`groupby("Type1")` هيجمع كل نوع في مجموعة لوحده،  
وبعدين نقدر نحسب متوسط أو مجموع كل مجموعة.

## Data Cleaning

Data cleaning is the process of fixing or removing incomplete, incorrect, or irrelevant data.

It is an essential step in data analysis because real-world data is often messy or missing values.

In fact, around **75% of the work done using Pandas is data cleaning**.


Data cleaning يعني تنظيف البيانات عن طريق إزالة أو تصحيح البيانات الغلط أو الناقصة.

وهي من أهم خطوات تحليل البيانات باستخدام Pandas، لأنها بتاخد معظم وقت العمل (حوالي 75%).

##### 1- Drop Irrelevant Columns

We remove columns that are not useful for analysis (irrelevant columns) to simplify the dataset and improve performance.

````python
print(df)
````

````python
df_temp = df.copy()
# عمل نسخة من الـ DataFrame الأصلي حتى لا يتم التأثير عليه

df_temp = df_temp.drop(columns=["Type2"])
# حذف العمود Type2 من النسخة فقط

print(df_temp)
# عرض البيانات بعد حذف العمود في النسخة
````

***********************

#### 2- Handle Missing Data

Missing data are empty or null values in the dataset. We handle them by either removing or filling them to ensure accurate analysis.

````python
print("=== BEFORE DROPNA ===")
print(df)

df_after_drop = df.dropna(subset=["Type2"])
# حذف الصفوف التي تحتوي على قيم مفقودة في Type2

print("\n=== AFTER DROPNA ===")
print(df_after_drop)
````

````python
print("=== BEFORE FILLNA ===")
print(df)

df_filled = df.fillna({"Type2": "None"})
# تعويض القيم المفقودة في Type2 بكلمة None

print("\n=== AFTER FILLNA ===")
print(df_filled)
````

********************************

#### 3- Fix Inconsistent Values

Inconsistent values are the same data written in different ways (like different capital letters or extra spaces). We fix them to make the data consistent and easier to analyze.

````python
df["Type1"] = df["Type1"].replace({"Grass": "GRASS"})
# توحيد القيم داخل العمود Type1

print(df["Type1"])
# طباعة العمود بعد التعديل للتأكد من التغيير
````

************************************

#### 4- Standardize Text

Standardizing text means converting text data into a consistent format.

This includes making all text lowercase or uppercase, and removing extra spaces to ensure uniformity in the dataset.

````python
df["Name"] = df["Name"].str.lower()
# تحويل كل النصوص في عمود Name إلى حروف صغيرة (lowercase)

print(df["Name"])
# طباعة العمود بعد التعديل
````



Pandas provides useful string methods to standardize text data:

- `str.lower()` → تحويل النص إلى حروف صغيرة
- `str.upper()` → تحويل النص إلى حروف كبيرة
- `str.title()` → تحويل أول حرف من كل كلمة إلى كبير

These methods help in cleaning and standardizing text data for consistent analysis.

## 5- Fix Data Types

Fixing data types means converting columns to the correct type (such as integer, float, or string).

This is important because wrong data types can lead to errors or incorrect calculations during analysis.

````python
df["Height"] = df["Height"].astype(float)
# تحويل عمود Height إلى نوع float
````

````python
df["Weight"] = df["Weight"].astype(int)
# تحويل عمود Weight إلى نوع int
````

###### ####astype() تُستخدم لتغيير نوع البيانات في الأعمدة.

**********

#### 6- Remove Duplicates

Removing duplicates means deleting repeated rows in the dataset.

Duplicate data can affect analysis results, so we remove them to ensure accuracy and consistency.

````python
df = df.drop_duplicates()
# حذف الصفوف المكررة من الـ DataFrame

print(df)
# عرض البيانات بعد إزالة التكرار
````
