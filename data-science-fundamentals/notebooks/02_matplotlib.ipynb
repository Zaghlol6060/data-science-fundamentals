## 🎨 Plot Customization 

في Matplotlib نقدر نعدل شكل الرسم البياني عشان يبقى أوضح وأجمل، وده بيشمل:

- Line Style (شكل الخط)
- Color (اللون)
- Markers (شكل النقاط)
- Size (حجم النقاط)

````python
x= np.array([2022,2023,2024,2025,2026])
y1=np.array([20,40,50,60,80])
y2=np.array([10,70,50,65,40])
y3=np.array([30,40,70,50,60])
````

````python
plt.plot(x,y,marker="1")
plt.show
````

### 📌 Reference (Markers List)

لو عايز تشوف كل أشكال الـ markers المتاحة:

👉 https://matplotlib.org/stable/api/markers_api.html


````python
plt.plot(x,y,marker="1",
            ms=20,
            mfc="#bd24bf",
            mec="#bd24bf")
plt.show
````

### 🧠 شرح الاختصارات

- `marker="1"` → شكل النقطة (Marker Style)  
- `ms = 20` → حجم النقطة (Marker Size)  
- `mfc = "#bd24bf"` → لون النقطة من الداخل (Marker Face Color)  
- `mec = "#bd24bf"` → لون حدود النقطة (Marker Edge Color)  

---

## 🎯 ملاحظات مهمة

- `mfc` = لون الداخل  
- `mec` = لون الإطار  
- لازم استخدام `plt.show()` لعرض الرسم  

---

````python
plt.plot(x,y,marker="o",
            ms=10,
            mfc="#bd24bf",
            mec="#bd24bf",
            linestyle="dashed",
            linewidth=3,
            color="#24bfa3")
plt.show
````

## 📌 خصائص الخط

- `linestyle="dashed"` → شكل الخط (متقطع)  
- `linewidth = 3` → سمك الخط  
- `color = "#24b24a"` → لون الخط  

---

## 📏 أنواع LineStyle في Matplotlib

- `'-'` → خط عادي (Solid line)  
- `'--'` → خط متقطع (Dashed line)  
- `'-.'` → خط شرطات ونقاط (Dash-dot)  
- `':'` → خط نقطي (Dotted line)  



**************************************

## 🎨 Using Dictionary for Plot Styling in Matplotlib

في المثال ده بنستخدم **dictionary** لتجميع كل خصائص شكل الخط مرة واحدة، بدل ما نكررها في كل plot.


````python
line_style = dict(
            marker="o",
            ms=10,
            mfc="#bd24bf",
            mec="#bd24bf",
            linestyle="-",
            linewidth=3,
                 )
````

````python
plt.plot(x,y1,color="#24bfa3",**line_style)
plt.plot(x,y2,color="#246fbf",**line_style)
plt.plot(x,y3,color="#6abf24",**line_style)
plt.show()

````

````python
line_style = dict(
            marker="o",
            ms=10,
            mfc="#bd24bf",
            mec="#bd24bf",
            linewidth=3,
                 )
````

````python
plt.plot(x,y1,color="#24bfa3",linestyle="-",**line_style)
plt.plot(x,y2,color="#246fbf",linestyle="--",**line_style)
plt.plot(x,y3,color="#6abf24",linestyle="-.",**line_style)
plt.show()
````

### 🧠 شرح الفكرة

بدل ما نكتب نفس الخصائص (مثل marker, linestyle, linewidth) لكل رسم، نقوم بتخزينها داخل dictionary.

ثم نستخدم `**` لتمرير هذه القيم مباشرة داخل دالة `plt.plot()`.

# Pie Charts

الـ Pie Chart هو رسم دائري بيتم تقسيمه إلى أجزاء (Slices) لتمثيل نسبة كل فئة من الإجمالي.

---

###  الفكرة

كل جزء من الدائرة يمثل نسبة (Percentage) لفئة معينة من البيانات.

---

###  أهميتها

- توضيح توزيع البيانات بين الفئات  
- عرض النسب بشكل بصري سهل  
- مفيد عند مقارنة أجزاء من الكل  

---

###  الاستخدام

نستخدم Pie Chart لما نكون عايزين نظهر:
- نسبة كل فئة من الإجمالي  
- توزيع البيانات بين Categories  





````python
catogories = np.array(["Grains","fruit","vegetables","protein"])
values = np.array([300,250,275,225])
color = ['#137046','#acc433','#c43363','#aa33c4']

plt.pie(values, labels=catogories, colors=color,autopct="%0.1f%%",explode=[0,0,0,0.1],shadow=True,startangle=180)
plt.title("Daily consuption",fontsize=18, family="Arial", fontweight="bold", color="#40b9c2")
plt.show()
````

### 📊 عرض النسب

- `autopct="%0.1f%%"`  
  → عرض النسبة المئوية داخل كل جزء  
  → رقم عشري واحد (مثال: 25.0%)

---

### 🎯 إبراز جزء معين

- `explode=[0,0,0,0.1]`  
  → إبعاد آخر جزء (protein) عن الدائرة لتمييزه  

---

### 🌑 الظل

- `shadow=True`  
  → إضافة ظل للشكل لزيادة الجمالية  

---

### 🔄 زاوية البداية

- `startangle=180`  
  → تدوير الرسم لبدء العرض من زاوية مختلفة  

---

# 📊 Scatter Plot

الـ Scatter Plot يُستخدم لعرض العلاقة بين متغيرين (Variables) عن طريق نقاط على الرسم.

---

###  الفكرة

كل نقطة في الرسم تمثل قيمة بين:
- متغير على المحور X  
- ومتغير على المحور Y  

---

###  أهميته

- تحديد العلاقة بين المتغيرات  
- معرفة نوع الارتباط (Correlation):
  - موجب (+) → كلما زاد X زاد Y  
  - سالب (-) → كلما زاد X قل Y  
  - لا يوجد ارتباط  

---

### مثال

- عدد ساعات المذاكرة (Study Hours)  
- درجات الاختبار (Test Scores)  


````python
x1=[0,1,1,2,3,4,5,6,7,7,8]  # hours study
y1=[55,60,65,62,68,70,75,77,79,80,85]  # grades

plt.scatter(x1,y1,color='#4CAF50',alpha=0.7,label="Class A")

x2=[0,1,2,2,3,4,5,6,7,8,9]  # hours study
y2=[50,55,58,60,63,67,69,72,74,76,78]  # different grades

plt.scatter(x2,y2,color='#FF5733',alpha=0.7,label="Class B")

plt.title("Study Hours vs Test Scores",fontsize=18, family="Arial", fontweight="bold", color="#333333")
plt.xlabel("Hours Studied",fontsize=14, fontweight="bold")
plt.ylabel("Grades",fontsize=14, fontweight="bold")

plt.legend()
plt.grid(alpha=0.3)

plt.show()
````

# شرح الفكرة

كل نقطة في الرسم بتمثل:
- عدد ساعات المذاكرة (X-axis)  
- درجة الطالب (Y-axis)  

---

##  المجموعات

- **Class A** → ممثلة باللون الأخضر  
- **Class B** → ممثلة باللون الأحمر  

كل مجموعة لها بيانات مختلفة علشان نقارن بينهم.

---

##  الخصائص المستخدمة

- `color` → تحديد لون كل مجموعة  
- `alpha` → شفافية النقاط (كل ما يقل الرقم تبقى أفتح)  
- `label` → اسم المجموعة (بيظهر في الـ legend)  

---

##  العنوان والمحاور

- `title` → عنوان الرسم (Study Hours vs Test Scores)  
- `xlabel` → عدد ساعات المذاكرة  
- `ylabel` → الدرجات  

---

##  Legend

- `plt.legend()` → بيعرض أسماء المجموعات (Class A, Class B)

---

##  Grid

- `plt.grid()` → بيساعد في قراءة القيم بسهولة  

---

##  التحليل

من الرسم نلاحظ:
- كلما زادت ساعات المذاكرة → زادت الدرجات  
- يعني فيه علاقة **موجبة (Positive Correlation)**  

---

# 📊 المدرج التكراري (Histogram)

الـ Histogram هو تمثيل بصري لتوزيع البيانات الرقمية (الكمية).

فكرته:
- تقسيم البيانات إلى فترات اسمها **Bins (فئات/مدى)**  
- حساب عدد القيم التي تقع داخل كل فترة  

---

##  الفكرة الأساسية:
كل عمود في المدرج التكراري يمثل **مدى من القيم**، وارتفاع العمود يمثل عدد القيم داخل هذا المدى.

---

##  استخداماته:
- فهم توزيع البيانات  
- معرفة شكل البيانات (طبيعي / منحرف)  
- اكتشاف القيم الشاذة (Outliers)  
- تحليل البيانات بشكل بصري  

---

##  مثال:
البيانات:
50, 55, 60, 60, 62, 70, 75, 80, 85, 90  

الفئات:
- 50–59 → عدد القيم  
- 60–69 → عدد القيم  
- 70–79 → عدد القيم  
- 80–89 → عدد القيم  
- 90–100 → عدد القيم  

---


````python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10, color="#913f3f", edgecolor="black")

plt.title("Distribution of Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

plt.show()
````

## 🧠 شرح الكود

### 🎲 إنشاء البيانات
- `np.random.normal(loc=80, scale=10, size=100)`  
  → إنشاء 100 قيمة بشكل توزيع طبيعي (Normal Distribution)
  - `loc=80` → المتوسط  
  - `scale=10` → مقدار التشتت (الانحراف المعياري)

---

### 🎯 تحديد القيم

- `np.clip(scores, 0, 100)`  
  → يجعل القيم بين 0 و 100 فقط (زي درجات الطلاب)

---

## 📊 رسم الـ Histogram

- `bins=10` → تقسيم البيانات إلى 10 مجموعات  
- `color` → لون الأعمدة  
- `edgecolor="black"` → إضافة حدود للأعمدة لتوضيح الشكل

# 📊 Subplots in Matplotlib

الـ Subplots هي طريقة لعرض أكثر من رسم بياني داخل نفس الشكل (Figure).

---

##  المفاهيم الأساسية

- **Figure** → هو الإطار الكبير (اللوحة الكاملة) التي تحتوي على الرسومات  
- **Axes (ax)** → هو كل رسم بياني منفصل داخل الـ Figure (Subplot)

---

##  الفكرة

بدل ما نرسم كل Graph لوحده، نقدر نجمع أكثر من رسم داخل نفس النافذة.

---

##  أهمية Subplots

- مقارنة أكثر من بيانات في نفس الوقت  
- تنظيم الرسومات بشكل احترافي  
- توفير مساحة وعرض أوضح  

---

##  مثال الاستخدام

نستخدم Subplots لما نريد:
- مقارنة Classes مختلفة  
- عرض Histogram + Line Plot معًا  
- تحليل أكثر من متغير في نفس العرض  

---

````python
print(plt.subplots(2,2))
````

````python
fig, axes = plt.subplots(2, 2)

x = np.array([0,1,1,2,3,4,5,6,7,7,8])

axes[0,0].plot(x, x*2, color="#ff5733")
axes[0,0].set_title("x * 2")

axes[0,1].plot(x, x**2, color="#33c1ff")
axes[0,1].set_title("x ** 2")

axes[1,0].plot(x, x**3, color="#33ff57")
axes[1,0].set_title("x ** 3")

axes[1,1].plot(x, np.arange(len(x)), color="#8e44ad")
axes[1,1].set_title("len(x)")

plt.tight_layout()
plt.show()
````


##  الفكرة

لدينا 4 رسومات مختلفة داخل نفس الشكل، وكل رسم يمثل عملية رياضية مختلفة على نفس البيانات.

---

##  شرح كل Subplot

- **subplot (0,0)** → يمثل `x * 2`  
- **subplot (0,1)** → يمثل `x ** 2`  
- **subplot (1,0)** → يمثل `x ** 3`  
- **subplot (1,1)** → يمثل طول البيانات `len(x)`

---


##  تنظيم الشكل

- `plt.subplots(2,2)` → ينشئ 4 رسومات داخل شبكة واحدة  
- `axes[i,j]` → يحدد مكان كل رسم داخل الشبكة  
- `plt.tight_layout()` → ينظم المسافات بين الرسومات
